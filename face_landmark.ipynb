```python
#face landmark

import mediapipe as mp
import cv2
```

```python
mp_detect=mp.solutions.face_detection
mp_drwing=mp.solutions.drawing_utils
```

```python
face_detect=mp_detect.FaceDetection()
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    res=face_detect.process(img)
    print(res.detections)
    if res.detections:
        for x in res.detections:
            mp_drwing.draw_detection(img,x)

    cv2.imshow("Face",img)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
